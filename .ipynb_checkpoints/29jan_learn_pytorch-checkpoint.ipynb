### Gradient Descent

```python
# f = w * x
# f = 2 * x

X = np.array([1, 2, 3, 4], dtype=np.float32)
Y = np.array([2, 4, 6, 8], dtype=np.float32)

w = 0.0

# model prediction
def forward(x):
  return w * x

# loss = MSE
def loss(y_hat, y):
  return ((y_hat - y)**2).mean()

# gradient
# MSE = 1/N * (w*x - y)**2
# dJ/dw = 2/N * (w*x - y) * x

def gradient(x, y, y_hat):
  return np.dot(2*x, y_hat-y).mean()

print(f'prediction before training: f(5) = {forward(5)}')

# training
lr = 0.01
n_iter = 20

for epoch in range(n_iter):
  # forward pass and calculate loss
  y_hat = forward(X)
  l = loss(y_hat, Y)

  # gradient
  dw = gradient(X, Y, y_hat)

  # update weights
  w -= lr * dw

  if epoch % 2 == 0:
    print(f'epoch: {epoch+1}, w: {w}, loss: {l}')


print(f'prediction after training: f(5) = {forward(5):.3f}')
```

```python
# f = w * x
# f = 2 * x

X = torch.tensor([1, 2, 3, 4], dtype=torch.float32)
Y = torch.tensor([2, 4, 6, 8], dtype=torch.float32)

w = torch.tensor(0.0, requires_grad=True)

# model prediction
def forward(x):
  return w * x

# loss = MSE
def loss(y_hat, y):
  return ((y_hat - y)**2).mean()

print(f'prediction before training: f(5) = {forward(5)}')

# training
lr = 0.01
n_iter = 100

for epoch in range(n_iter):
  # forward pass and calculate loss
  y_hat = forward(X)
  l = loss(y_hat, Y)

  # gradient
  l.backward() # dl/dw
  dw = w.grad

  # update weights
  with torch.no_grad():
    w -= lr * dw

  # zero gradient
  w.grad.zero_()

  if epoch % 10 == 0:
    print(f'epoch: {epoch+1}, w: {w}, loss: {l}')


print(f'prediction after training: f(5) = {forward(5):.3f}')
```

### Training Pipeline

```python
# 1. Design model (input, output size, forward pass)
# 2. Construct loss and optimizer
# 3. Training loop
#     - forward pass : compute prediction
#     - backward pass : gradients
#     - update weights
```

```python
# f = w * x
# f = 2 * x

import torch.nn as nn

X = torch.tensor([1, 2, 3, 4], dtype=torch.float32)
Y = torch.tensor([2, 4, 6, 8], dtype=torch.float32)

w = torch.tensor(0.0, requires_grad=True)

# model prediction
def forward(x):
  return w * x

loss = nn.MSELoss()
optimizer = torch.optim.SGD([w], lr=0.01)

print(f'prediction before training: f(5) = {forward(5)}')

# training
n_iter = 100

for epoch in range(n_iter):
  # forward pass and calculate loss
  y_hat = forward(X)
  l = loss(Y, y_hat)

  # gradient
  l.backward() # dl/dw

  # update weights
  optimizer.step()

  # zero gradient
  optimizer.zero_grad()

  if epoch % 10 == 0:
    print(f'epoch: {epoch+1}, w: {w}, loss: {l}')


print(f'prediction after training: f(5) = {forward(5):.3f}')
```

```python
import torch
import torch.nn as nn
import numpy as np
from sklearn import datasets
import matplotlib.pyplot as plt
```

```python
# 0. prepare datasets
X_numpy, Y_numpy = datasets.make_regression(n_samples=100, n_features=1, noise=20, random_state=1)
X = torch.from_numpy(X_numpy.astype(np.float32))
Y = torch.from_numpy(Y_numpy.astype(np.float32))
```

```python
n_samples, n_features = X.shape
input_size, output_size = n_features, n_features
```

```python
# 1. model
model = nn.Linear(input_size, output_size)

# 2. loss and optimizer
lr = 0.01
loss = nn.MSELoss()
optimizer = torch.optim.SGD(model.parameters(), lr)

# 3. training loop
num_epochs = 1000
for epoch in range(num_epochs):
  # forward pass and calculate loss
  Y_hat = model(X)
  l = loss(Y_hat, Y)

  # backward pass
  l.backward() # dl/dw
  optimizer.step() # update weights
  optimizer.zero_grad() # reset gradient

  if epoch % 100 == 0:
    print(f'iteration = {epoch+1}, loss = {l}')
```

### Logistic Regression

```python
import torch 
import torch.nn as nn
import numpy as np
from sklearn import datasets
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
bc = datasets.load_breast_cancer()
X, Y = bc['data'], bc['target']
```

```python
X.shape
```

```python
Y.shape
```

```python
Y = Y.reshape(Y.shape[0], 1)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=32)
```

```python
X_train[0]
```

```python
sc = StandardScaler()
```

```python
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
X_train[0]
```

```python
# convert into tensor
X_train = torch.from_numpy(X_train.astype(np.float32))
X_test = torch.from_numpy(X_test.astype(np.float32))
Y_train = torch.from_numpy(Y_train.astype(np.float32))
Y_test = torch.from_numpy(Y_test.astype(np.float32))
```

```python
X_train
```

```python
rows, cols = X_train.shape
cols
```

```python
y_rows, y_cols = Y_train.shape
y_cols
```

```python
# design model
class LogisticRegression(nn.Module):
    def __init__(self, input_dim, output_dim):
        super(LogisticRegression, self).__init__()
        # define layers
        self.lin = nn.Linear(input_dim, output_dim)

    def forward(self, x):
        y = self.lin(x)
        # return 1/(1 + torch.exp(-y))
        return torch.sigmoid(y)

model = LogisticRegression(cols, y_cols)

# create loss and optimizer
lr = 0.01
loss = nn.MSELoss()
optimizer = torch.optim.SGD(model.parameters(), lr)


num_epochs = 100
for epoch in range(num_epochs):
    # forward and compute loss
    Y_pred = model(X_train)
    l = loss(Y_train, Y_pred)

    # backward
    l.backward() # dl/dw
    optimizer.step() # update w -= lr * dw
    optimizer.zero_grad()

    if epoch % 5 == 0:
        print(f'epoch = {epoch+1}, loss = {l}')
```

```python
with torch.no_grad(): # so the gradient is not used by the computation
    Y_hat = model(X_test).round()

plt.plot(X_test, Y_test, 'ro')
plt.plot(X_test, Y_hat, 'bo');
```

```python
# calculating accuracy
acc = np.int32((Y_hat == Y_test).sum()) / len(Y_hat)
acc
```

### Dataset Transform

```python
dataset = torchvision.datasets.MNIST(
    root='./data', transform=torchvision.transforms.ToTensor(),
)
```

```python
class WineDataset(Dataset):
    def __init__(self, transform=None):
        xy = np.loadtxt('wine.csv', delimiter=',', dtype=np.float32, skiprows=1)

        self.n_samples = xy.shape[0]
        self.x = xy[:, 1:]
        self.y = xy[:, [0]]
        self.transform = transform

    def __getitem__(self, index):
        sample = self.x[index], self.y[index]
        if self.transform:
            sample = self.transform(sample)
        return sample
        
    def __len__(self):
        return self.n_samples
```

```python
class ToTensor:
    # call method
    def __call__(self, sample):
        inputs, targets = sample
        return torch.from_numpy(inputs), torch.from_numpy(targets)
```

```python
class MulTransform:
    def __init__(self, factor):
        self.factor = factor

    def __call__(self, sample):
        inputs, targets = sample
        inputs *= self.factor
        return inputs, targets
```

```python
dataset = WineDataset(transform=ToTensor())
```

### Feed Forward

```python
import torch
import torch.nn as nn
import torchvision
import torchvision.transforms as transforms
import matplotlib.pyplot as plt
from torch.utils.data import DataLoader
```

```python
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
```

```python
device
```

```python
input_size = 28*28
hidden_size = 100
num_classes = 10
num_epochs = 3
batch_size = 32
learning_rate = 0.01

# dataset
train_dataset = torchvision.datasets.MNIST(root='./data', train=True, transform=transforms.ToTensor())
test_dataset = torchvision.datasets.MNIST(root='./data', train=False, transform=transforms.ToTensor())
```

```python
train_loader = DataLoader(dataset=train_dataset, batch_size=batch_size, shuffle=True)
test_loader = DataLoader(dataset=test_dataset, batch_size=batch_size, shuffle=False)
```

```python
examples = iter(train_loader)
sample, label = next(examples)
```

```python
sample.shape, label.shape
```

```python
label
```

```python
for i in range(6):
    plt.subplot(2, 3, i+1)
    plt.imshow(sample[i][0])
```

```python
# model

class NeuralNet(nn.Module):
    def __init__(self, input_size, hidden_size, num_classes):
        super(NeuralNet, self).__init__()
        # layers
        self.linear1 = nn.Linear(input_size, hidden_size)
        self.relu = nn.ReLU()
        self.linear2 = nn.Linear(hidden_size, num_classes)

    def forward(self, x):
        out = self.linear1(x)
        out = self.relu(out)
        out = self.linear2(out)
        return out

model = NeuralNet(input_size=input_size, hidden_size=hidden_size, num_classes=num_classes)
```

```python
# loss and optimizer
optimizer = torch.optim.Adam(model.parameters(), lr=learning_rate)
loss = nn.CrossEntropyLoss()
```

```python
# training
for epoch in range(num_epochs):
    for i, (images, labels) in enumerate(train_loader):
        # 32, 1, 28, 28 -> 32, 784
        images = images.reshape(-1, 28*28).to(device)
        labels = labels.to(device)
        # forward and compute loss
        Y_pred = model(images)
        l = loss(Y_pred, labels)
        # backward
        l.backward() # dl/dw
        optimizer.step() # update weights w -= lr * dw
        optimizer.zero_grad()

    print(f'epoch = {epoch+1}, loss={l}')
```

```python
with torch.no_grad():
    n_correct = 0
    n_samples = 0
    for images, labels in test_loader:
        images = images.reshape(-1, 28*28).to(device)
        labels = labels.to(device)
        Y_pred = model(images)
        # value, index
        _, pred = torch.max(Y_pred, 1)
        n_samples += labels.shape[0]
        n_correct += (pred == labels).sum()

acc = n_correct/n_samples
print(acc)
```
